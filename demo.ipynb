```python
from backpropogation_engine import value 
from neural_network import MLP 
from graphing_tool import draw_connections 
```

```python
xi = [ 
    [2.0 , -3.0 , 5.0] ,
    [5.0 , 11.0 , -12.0] ,
    [23.0 , 1.0 , -1.0] ,
    [0.5 , 25.0 , -8.0] ,
] # inputs to the neural network 

yo = [ 1.0 , -1.0 , 1.0 , -1.0 ] #desired outputs

n = MLP(3, [4, 4, 1]) # defining the size of neural network

print(f"Total parameters in the network : {len(n.parameters())} ")

for k in range(50) : 

    # forward pass 
    ypred = [n(x) for x in xi]

    loss = sum ((yout-yint)**2 for yout , yint in zip(ypred,yo)) 

    # backward pass 
    n.zero_grad() 
    loss.back_propogation() 

    # update 
    for p in n.parameters() : 
        p.data -= 0.2 * p.grad 

    print(k,loss.data)

print("-----FINAL RESULTS AFTER FINE TUNING THE PARAMETERS------")
print(ypred)

print("-----GRAPHING THE BACKPROPOGATION OF LOSS ------")

draw_connections(loss) 
```
